# StarDist Prediction Test Notebook

Test a trained StarDist model on images 10-15.

**Prerequisites:** Run `test_stardist_training.ipynb` first.

## 1. Setup

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from skimage.segmentation import find_boundaries

from napari_ai_lab.models.image_data_model import ImageDataModel
from napari_ai_lab.Segmenters.GlobalSegmenters.StardistSegmenter import (
    StardistSegmenter,
)
```

## 2. Load Project & Model

```python
# Load project (same as launch_nd_ai_lab.py pattern)
project_path = Path("../tests/test_images/nuclei").resolve()
model = ImageDataModel(str(project_path))

model.set_annotation_io_type(
        "stacked_sequence"
)

model.set_prediction_io_type(
    "stacked_sequence"
)

# Load all images as a stack
images = model.load_image(0, stacked=True)
#images = model.load_image(0)
print(f"Loaded image stack: {images.shape}")

# Load trained StarDist model
model_name = "stardist_nuclei_test"
segmenter = StardistSegmenter()
segmenter.model_save_dir = str(project_path / "models")
segmenter.set_model(model_name)
print(f"Loaded model: {model_name}")
```

## 3. Predict on Range (Images 10-15)

```python
# Define range (0-based indexing)
start_idx = min(9, len(images) - 1)   # Image 10
end_idx = min(14, len(images) - 1)    # Image 15

# Storage for results
predictions = {}

def on_done(step, mask):
    idx = step[0]
    print('current step:', step)
    predictions[idx] = mask
    print(f"  ✓ Image {idx + 1}: {len(np.unique(mask)) - 1} objects")
    model.save_predictions(
                    mask,
                    0,
                    current_step=step,
                    selected_axis="YX",
                )

# Run batch prediction (synchronous)
print(f"Predicting images {start_idx + 1} to {end_idx + 1}...")
model.segment_range(
    segmenter=segmenter,
    selected_axis="YX",
    start_index=start_idx,
    end_index=end_idx,
    on_slice_done=on_done,
    use_threading=False,
)
print(f"✅ Done! Predicted {len(predictions)} images.")
```

## 4. Visualize Results

```python
# Display input images and predictions
n = len(predictions)
fig, axes = plt.subplots(2, n, figsize=(4 * n, 8))
if n == 1:
    axes = axes.reshape(2, 1)

for col, (idx, pred) in enumerate(sorted(predictions.items())):
    img = images[idx]
    n_obj = len(np.unique(pred)) - 1

    axes[0, col].imshow(img, cmap="gray")
    axes[0, col].set_title(f"Image {idx + 1}")
    axes[0, col].axis("off")

    axes[1, col].imshow(pred, cmap="nipy_spectral", interpolation="nearest")
    axes[1, col].set_title(f"{n_obj} objects")
    axes[1, col].axis("off")

plt.tight_layout()
plt.show()
```

## 5. Show Contour Overlays

```python
# Display overlays
n = len(predictions)
fig, axes = plt.subplots(1, n, figsize=(6 * n, 6))
if n == 1:
    axes = [axes]

for col, (idx, pred) in enumerate(sorted(predictions.items())):
    img = images[idx]
    bounds = find_boundaries(pred, mode="outer")

    axes[col].imshow(img, cmap="gray")
    axes[col].imshow(np.ma.masked_where(~bounds, bounds), cmap="spring", alpha=0.8)
    axes[col].set_title(f"Image {idx + 1}")
    axes[col].axis("off")

plt.tight_layout()
plt.show()
```

# StarDist Prediction Test Notebook

This notebook demonstrates batch prediction using a trained StarDist model on a range of images.

**Prerequisites:** Run `test_stardist_training.ipynb` first to train the model.

**Workflow:**
1. **Setup** — load the nuclei project and trained model
2. **Load image range** — prepare images 10-15 for prediction
3. **Batch predict** — use `segment_range()` to predict on all images
4. **Visualize results** — display all predictions side-by-side

## 1. Import Libraries

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import tifffile

from napari_ai_lab.models.image_data_model import ImageDataModel
from napari_ai_lab.Segmenters.GlobalSegmenters.StardistSegmenter import (
    StardistSegmenter,
)
```

## 2. Setup Project and Load Trained Model

We'll use the same nuclei project and load the model trained in `test_stardist_training.ipynb`.

```python
# Use relative path from notebook location to test images
project_path = Path("../tests/test_images/nuclei").resolve()

# Model details (must match test_stardist_training.ipynb)
model_name = "stardist_nuclei_test"
model_base_path = project_path / "models"
model_full_path = model_base_path / model_name

print(f"Project: {project_path}")
print(f"Model path: {model_full_path}")
print(f"Model exists: {model_full_path.exists()}")

if not model_full_path.exists():
    raise FileNotFoundError(
        f"Model not found at {model_full_path}. "
        "Please run test_stardist_training.ipynb first to train the model."
    )
```

## 3. Create Segmenter and Load Trained Model

```python
# Create segmenter and point it to our trained model
segmenter = StardistSegmenter()
segmenter.model_save_dir = str(model_base_path)
segmenter.set_model(model_name)

print(f"Loaded model: {segmenter.inference_model_name}")
print(f"Model axis: {segmenter.get_model_axis_map().get(model_name, 'YX')}")
```

## 4. Load Images as Stack

We'll load all the nuclei images and create a 3D stack (N, Y, X) where N is the image index.
Then we'll use `ImageDataModel.segment_range()` to predict on images 10-15.

```python
# Load all images into a stack
image_files = sorted(project_path.glob("*.tif"))
print(f"Found {len(image_files)} images")

# Load images into a 3D numpy array (N, Y, X)
images_list = []
for img_path in image_files:
    img = tifffile.imread(str(img_path))
    images_list.append(img)

# Stack into 3D array
images_stack = np.array(images_list)
print(f"Stack shape: {images_stack.shape} (N, Y, X)")

# Define the range we want to predict (images 10-15)
# Note: Use 0-based indexing, so image 10 is index 9
start_idx = 9  # Image 10 (0-based)
end_idx = 14   # Image 15 (0-based, inclusive)

# Clamp to available images
start_idx = min(start_idx, len(images_stack) - 1)
end_idx = min(end_idx, len(images_stack) - 1)

if start_idx > end_idx:
    raise ValueError(
        f"Not enough images. Found {len(images_stack)} images, "
        f"need at least {end_idx + 1} to test range 10-15."
    )

print(f"\nWill predict on images {start_idx + 1} to {end_idx + 1} (1-based)")
print(f"That's {end_idx - start_idx + 1} images total")
```

## 5. Setup ImageDataModel with Stacked Images

We'll create an `ImageDataModel` and assign the image stack, then use `segment_range()` to process images 10-15.

```python
# Create ImageDataModel
image_data_model = ImageDataModel(str(project_path))

# Assign the stacked images (this is what napari would do when loading a stack)
image_data_model.image_data = images_stack
image_data_model.axis_types = "TYX"  # T for "image index", Y, X for spatial

print(f"ImageDataModel ready with shape: {image_data_model.image_data.shape}")
```

## 6. Batch Predict on Range (Synchronous)

Use the new `segment_range()` method to predict on all images in the range.
We'll collect results in a dictionary for later visualization.

```python
# Storage for results
predictions = {}

def on_slice_done(current_step, mask):
    """Callback to store each prediction result."""
    slice_idx = current_step[0]  # First dimension is the image index
    predictions[slice_idx] = mask
    print(f"  ✓ Image {slice_idx + 1}: {mask.shape}, {len(np.unique(mask)) - 1} objects")

def on_progress(current, total):
    """Progress callback."""
    print(f"Progress: {current}/{total}")

# Run batch prediction (synchronous, no threading for notebook simplicity)
print("Starting batch prediction...")
processor = image_data_model.segment_range(
    segmenter=segmenter,
    selected_axis="YX",  # 2D spatial segmentation
    start_index=start_idx,
    end_index=end_idx,
    axes_to_collapse=None,
    on_slice_done=on_slice_done,
    on_progress=on_progress,
    use_threading=False,  # Synchronous for notebooks
)

print(f"\n✅ Batch prediction complete! Processed {len(predictions)} images.")
```

## 7. Visualize All Predictions

Display all input images and their predictions side-by-side.

```python
# Create a grid showing all images and their predictions
n_images = len(predictions)
fig, axes = plt.subplots(2, n_images, figsize=(4 * n_images, 8))

if n_images == 1:
    axes = axes.reshape(2, 1)

for col_idx, (img_idx, pred) in enumerate(sorted(predictions.items())):
    # Get the original image
    img = images_stack[img_idx]

    # Top row: input image
    axes[0, col_idx].imshow(img, cmap="gray")
    axes[0, col_idx].set_title(f"Image {img_idx + 1}")
    axes[0, col_idx].axis("off")

    # Bottom row: prediction
    n_objects = len(np.unique(pred)) - 1
    axes[1, col_idx].imshow(pred, cmap="nipy_spectral", interpolation="nearest")
    axes[1, col_idx].set_title(f"Prediction\n{n_objects} objects")
    axes[1, col_idx].axis("off")

plt.suptitle(f"StarDist Predictions: Images {start_idx + 1}-{end_idx + 1}", fontsize=14)
plt.tight_layout()
plt.show()
```

## 8. Detailed View with Overlays

Show prediction contours overlaid on the input images.

```python
from skimage.segmentation import find_boundaries

# Create overlays for all predictions
n_images = len(predictions)
fig, axes = plt.subplots(1, n_images, figsize=(6 * n_images, 6))

if n_images == 1:
    axes = [axes]

for col_idx, (img_idx, pred) in enumerate(sorted(predictions.items())):
    # Get the original image
    img = images_stack[img_idx]

    # Find boundaries
    boundaries = find_boundaries(pred, mode="outer")

    # Display overlay
    axes[col_idx].imshow(img, cmap="gray")
    axes[col_idx].imshow(
        np.ma.masked_where(~boundaries, boundaries),
        cmap="spring",
        alpha=0.8
    )
    n_objects = len(np.unique(pred)) - 1
    axes[col_idx].set_title(f"Image {img_idx + 1}\n{n_objects} objects")
    axes[col_idx].axis("off")

plt.suptitle("Input Images with StarDist Contours", fontsize=14)
plt.tight_layout()
plt.show()
```

## 9. Summary Statistics

Show object counts and other metrics for all predictions.

```python
import pandas as pd

# Collect statistics
stats = []
for img_idx in sorted(predictions.keys()):
    pred = predictions[img_idx]
    n_objects = len(np.unique(pred)) - 1  # Exclude background

    # Calculate mean object size
    if n_objects > 0:
        object_sizes = [np.sum(pred == label) for label in range(1, n_objects + 1)]
        mean_size = np.mean(object_sizes)
        std_size = np.std(object_sizes)
    else:
        mean_size = 0
        std_size = 0

    stats.append({
        'Image': img_idx + 1,
        'Objects': n_objects,
        'Mean Size (px)': f"{mean_size:.1f}",
        'Std Size (px)': f"{std_size:.1f}",
    })

# Display as table
df_stats = pd.DataFrame(stats)
print("\n📊 Prediction Statistics:")
print(df_stats.to_string(index=False))

# Overall summary
total_objects = df_stats['Objects'].sum()
print(f"\n✅ Total objects detected across all {len(predictions)} images: {total_objects}")
```
